```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder


import pandas as pd
import numpy
import tensorflow as tf

numpy.random.seed(3)
tf.random.set_seed(3)

#데이터
df = pd.read_csv('sonar.csv',header = None) #header = none이 무슨 뜻이지

dataset = df.values
X=dataset[:,0:60]
Y_obj = dataset[:,60]

#문자열 변환
e = LabelEncoder()
e.fit(Y_obj) #이부분 다시 봐야될듯
Y = e.transform(Y_obj)

#모델설정
model = Sequential()
model.add(Dense(24,input_dim=60,activation='relu'))
model.add(Dense(10,activation = 'relu')) #relu 뭐였는지 기억안나,,
model.add(Dense(1,activation = 'sigmoid'))
#모델 컴파일
model.compile(loss = 'mean_squared_error',optimizer = 'adam', metrics = ['accuracy'])
#모델 실행
model.fit(X,Y,epochs = 200, batch_size = 5)
#r결과
print("\n Accuracy : %.4f" % (model.evaluate(X,Y)[0])) #이부분도 다시 봐야될듯. model.evaluate이 뭐지
```

```python
#k겹 교차검증
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from keras.models import load_model
from sklearn.model_selection import StratifiedKFold

import pandas as pd
import numpy
import tensorflow as tf

seed = 0
numpy.random.seed(seed)
tf.random.set_seed(3)
#이부분 왜하는지 다시 봐야함.


#데이터
df = pd.read_csv('sonar.csv',header = None) #header = none이 무슨 뜻이지

dataset = df.values
X=dataset[:,0:60]
Y_obj = dataset[:,60]
print(type(dataset))
#문자열 변환
e = LabelEncoder()
e.fit(Y_obj) #이부분 다시 봐야될듯
Y = e.transform(Y_obj)
print(type(X))
print(type(Y))
#10개로 쪼개기
n_fold = 10
skf = StratifiedKFold(n_splits = n_fold, shuffle = True, random_state = seed)

#각각의 정확도를 저장하는 배열
accuracy = []

#모델에 k겹 교차검증을 적용하기때문에 모델이 바뀜
for train,test in skf.split(X,Y) :
    print(train)
    print(test)
    model = Sequential()

    model.add(Dense(24,input_dim=60,activation='relu'))
    model.add(Dense(10,activation = 'relu')) #relu 뭐였는지 기억안나,,
    model.add(Dense(1,activation = 'sigmoid'))
    model.compile(loss = 'mean_squared_error',optimizer = 'adam', metrics = ['accuracy'])

    model.fit(X[train],Y[train],epochs=130,batch_size=5)
              
    k_accuracy = "%.4f" %(model.evaluate(X[test],Y[test])[1])
    accuracy.append(k_accuracy)


#모델이 만족스러워서 모델 저장
model.save('my_model.h5')

#메모리 내의 모델 삭제
del model
#모델을 새로 불러옴.
model = load_model('my_model.h5')
#그럼 이거 저장된 모델이 어디로 가는거지....?? 메모리에 있는건가....? 

#테스트 셋에 모델 적용
print("\n %.f fold accuracy: " %n_fold,accuracy)
```
